```python
# Install libraries dependencies
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
from bs4 import BeautifulSoup
import json
import os
import re
from selenium.webdriver.firefox.options import Options
import csv
```

```python
# Retrieve Answers to Questions within CSV file
def retrieve(topic, genre):
    #create a subdirectory and put the file inside
    subdirectory = genre + " Questions"
    #check if subdirectory (genre) exists, if not create one
    if not os.path.exists(subdirectory):
        os.makedirs(subdirectory)
    # The file is saved within a parent directory that is labelled as your topic name with each file labelled as the topic name
    filename = re.sub(r'[<>:"/\\|?*]', '', topic) + '.csv'
    filename = os.path.join(subdirectory, filename)
    # Creates a csv file and iterates between each topic, depositing it in a dictionary and appending it into the jsonl file
    if os.path.exists(filename):
        print("File", filename, "already exists. Skipping.")
    else:
        with open(filename, 'w', encoding='utf-8') as csv_file:
            writer = csv.writer(csv_file)
            options = Options()
            # CHANGE THE LINK TO YOUR FIREFOX PROFILE WHERE QUORA IS LOGGED IN + ENABLE COOKIES
            options.profile = ''
            driver = webdriver.Firefox(options=options)
            driver.get("https://www.quora.com/search?q=" + topic.replace("-","") + "&type=question")
            time.sleep(5)
            # select how many times you want to scroll down the web page to retrieve the questions
            # Relevancy will be impacted the more you scroll
            for i in range(5):
                driver.execute_script("window.scrollTo(0, document.body.scrollHeight)")
                time.sleep(2)
            # Get the html source of the page
            html_source = driver.page_source
            soup_3 = BeautifulSoup(html_source, 'html.parser')
            # Find all elements that contains the questions using XPath
            xpathh = "(//div[contains(@class,'q-flex qu-flexDirection--row')])"
            elements = driver.find_elements(By.XPATH, value=xpathh)
            # Initialize the header of the csv file
            initalize = ["Keyword","Title","Quora_link"]
            writer.writerow(initalize)
            # Iterates through the questions and saves it within the csv file
            for element in elements:
                inner_html = element.get_attribute('innerHTML')
                soup = BeautifulSoup(inner_html, 'html.parser')
                plain_text = soup.get_text(separator=' ', strip=True)
                link = "https://www.quora.com/"
                # In the case that when retrieving the questions, each question has more than 1 question
                if plain_text.count("?") > 1:
                    plain_text = plain_text.split("?")
                    for i in plain_text:
                        # To get rid of the last index since it is empty
                        if len(i) == 0:
                            continue
                        else:
                            plain_text = i.strip()
                            plain_text += "?"
                            print(str(plain_text))
                            plain_text = plain_text.replace(".","")
                            link = "https://www.quora.com/" + plain_text.replace(" ","-")
                            things = [topic, plain_text , link]
                            writer.writerow(things)
                else:
                    plain_text = plain_text.rstrip("?")
                    plain_text = plain_text.strip()
                    plain_text += "?"
                    plain_text = plain_text.replace(". ","")
                    plain_text = plain_text.replace(" .","")
                    plain_text = plain_text.replace(".","")
                    print(str(plain_text))
                    link = "https://www.quora.com/" + plain_text.replace(" ","-")
                    things = [topic, plain_text , link]
                    writer.writerow(things)
            # Close the browser
            driver.quit()


    


```

```python
#Removes empty files due to error links within the subdirectory
def remove_empty_files(subdirectory):
    # Iterate over files in the subdirectory
    for filename in os.listdir(subdirectory):
        filepath = os.path.join(subdirectory, filename)
        # Check if the file is empty
        if os.path.isfile(filepath) and os.path.getsize(filepath) == 0:
            print("Removing empty file:", filename)
            os.remove(filepath)
```

```python
# Reads the file and retrieves the list of questions along side with the topic
# Questions CSV is formatted per row as (TOPICS)
with open("Put the topics you want to scrape here.csv") as topics:
    next(topics)
    for line in topics:
        line = line.rstrip("\n")
        line = line.split(",")
        topic = line[0]
        genre = line[1]

        #calls the function to find the answer to each question, the directory would be the topic
        retrieve(topic, genre)
    subdirectory = genre + " Questions"
    remove_empty_files(subdirectory)
        
```
